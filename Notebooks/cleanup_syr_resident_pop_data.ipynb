# Cleanup Syrian resident population data at administrative level 3
## Find data points in which district names are different to names from Syria shapefile
## Harmonise names manually

```python
import geopandas
import pandas as pd

df = geopandas.read_file("data/syria/syrian_districts_3.geojson")
df_names = df.ADM3_EN
df_names
```

```python
list1 = list(df_names)
list1
```

```python
# Extract names of Syrian districts into a dataframe
df2 = pd.DataFrame({'ADM3_EN':list1})
df2 = df2.astype('string')
df2
```

```python
df1 = pd.read_csv("data/syria/syr_pop_2004.csv", delimiter = ";")
df1 = df1.rename(columns={'SUB-DISTRICT Nahya English':'ADM3_EN'})
df1 = df1[['ADM3_EN', 'Total_population']]

#df1['ADM3_EN'] = df1['adm1_en'].astype('string')
#df1['ADM3_EN'] = df1['adm1_en'].str.upper()

# Change number format from xx.yyy to xxyyy
df1['Total_population'] = df1['Total_population'].astype('string')
df1['Total_population'] = df1['Total_population'].str.replace('.', '', regex=True)
df1
```

```python
# Harmonisation of province names across data sets
joined = df2.join(df1.set_index('ADM3_EN'), on='ADM3_EN')
joined
```

```python
joined_null = joined[joined['Total_population'].isnull()]
joined_null
```
