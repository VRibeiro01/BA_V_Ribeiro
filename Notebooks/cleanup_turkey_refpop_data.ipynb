```python
import geopandas
import pandas as pd

df = geopandas.read_file("data/turkey_districts_1.geojson")
df_names = df.adm1_en
df_names
```

```python
list1 = list(df_names)
list1
```

```python
# Extract names of Turkish provinces into a dataframe
df2 = pd.DataFrame({'adm1_en':list1})
df2 = df2.astype('string')
df2
```

```python
import pandas as pd
df1 = pd.read_csv("data/refpop_turkey_csv.csv", delimiter = ";")
df1['adm1_en'] = df1['adm1_en'].astype('string')
df1['adm1_en'] = df1['adm1_en'].str.upper()

# Change number format from xx.yyy to xxyyy
df1['Number'] = df1['Number'].astype('string')
df1['Number'] = df1['Number'].str.replace('.', '', regex=True)
df1
```

```python
# Harmonisation of province names across data sets
joined = df2.join(df1.set_index('adm1_en'), on='adm1_en')
joined
```

```python
joined_null = joined[joined['Number'].isnull()]
joined_null
```

```python
joined_not_null = joined[joined['Number'].notnull()]
joined_not_null = joined_not_null.rename(columns = {'adm1_en' : 'Region', 'Number': 'IDPs'})
joined_not_null.dtypes

```
