```python
import geopandas

df = geopandas.read_file("data/Border_Crossing_Points_SyrTur.geojson")

# Keep only relevant columns
cols = ['pcode', 'geometry']
df = df[cols]

# Unify coordinate system
df = df.to_crs('EPSG:4326')
df
```

```python
df1 = geopandas.read_file("data/turkey2_syria3_districts.geojson")
df1
```

```python

# Merge border crossing points coordinates with districts dataset
join_left_df = df.sjoin(df1, how="left")
join_left_df = join_left_df[join_left_df['index_right'].notnull()]
join_left_df
```

```python
# Keep only relevant columns
cols = ['geometry', 'ADM2_EN','ADM3_EN']
join_left_df = join_left_df[cols]
join_left_df
```

```python
# Show only Turkish districts that contain border crossing point

border_cross_syr = join_left_df[join_left_df['ADM3_EN'].isnull()]
border_cross_syr = border_cross_syr.drop_duplicates(subset='ADM2_EN', keep="first")
border_cross_syr = border_cross_syr.drop(columns=['ADM3_EN'])
border_cross_syr = border_cross_sry = border_cross_syr.rename(columns={'ADM2_EN':'Region'})
border_cross_syr
```
