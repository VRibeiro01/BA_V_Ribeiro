```python
import pandas as pd
df = pd.read_csv("data/conflicts_tur_syr_22.csv", delimiter = ",")
df

```

```python
# Split date into day and month
df["event_date"] = pd.to_datetime(df["event_date"])
df['day'] = df ['event_date'].dt.day 
df ['month'] = df ['event_date'].dt.month
df = df.drop("event_date", axis=1)
df
```

```python
# Extract relevant columns
important_cols = ["event_id_cnty", "day","month","year","location", "latitude", "longitude"]
interesting_data = df[important_cols]
interesting_data
```
